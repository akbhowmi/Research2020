# Original halo from the uniform simulation box

```python
path_to_output='/ufrc/lblecha/aklantbhowmick/arepo_runs_aklant/' # this is the folder containing the simulation run
run='L25n128MUSIC_rerun/' # name of the simulation runs
index_of_selected_halo=100 # index of the halo you want to select 
desired_redshift_of_selected_halo=0 #your desired redshift. Note that the code will choose the closest redshift available in the data 
basePath=path_to_output+run+'/output/' #your complete path to the output files
boxsize=arepo_package.get_box_size(basePath)
particle_property='Coordinates' #desired particle property
#--------------------------Choose dark matter particle type and read the particle properties----------------------------
p_type=1 
DM_particle_positions,output_redshift=arepo_package.get_particle_property_within_groups(basePath,particle_property,p_type,desired_redshift_of_selected_halo,index_of_selected_halo,group_type='groups',list_all=True)
#------------------------------------------------------------------------------------------------------------------------
#--------------------------Choose star matter particle type and read the particle properties----------------------------
p_type=4 
star_particle_positions,output_redshift=arepo_package.get_particle_property_within_groups(basePath,particle_property,p_type,desired_redshift_of_selected_halo,index_of_selected_halo,group_type='groups',list_all=True)
#------------------------------------------------------------------------------------------------------------------------
#--------------------------Choose black hole particle type and read the particle properties------------------------------
p_type=5 
bh_particle_positions,output_redshift=arepo_package.get_particle_property_within_groups(basePath,particle_property,p_type,desired_redshift_of_selected_halo,index_of_selected_halo,group_type='groups',list_all=True)
#------------------------------------------------------------------------------------------------------------------------

#--------------------------Plot the histograms of their positions--------------------------------------------------------
f,ax=plt.subplots(figsize=(10,10))
NBINS=30
ax.hist2d(DM_particle_positions[:,0],DM_particle_positions[:,1], bins=(NBINS,NBINS), norm=mpl.colors.LogNorm(),cmap='Blues_r');
ax.hist2d(star_particle_positions[:,0],star_particle_positions[:,1], bins=(NBINS,NBINS), norm=mpl.colors.LogNorm(),cmap='Reds_r');
ax.hist2d(DM_particle_positions[:,0],DM_particle_positions[:,1], bins=(NBINS,NBINS), norm=mpl.colors.LogNorm(),cmap='Blues_r',alpha=0);
ax.errorbar(bh_particle_positions[:,0],bh_particle_positions[:,1],marker='o',linewidth=0, color='black',ms=10);
ax.tick_params(labelsize=30)
ax.set_xlabel('x(kpc/h)',fontsize=30)
ax.set_ylabel('y(kpc/h)',fontsize=30)

```
